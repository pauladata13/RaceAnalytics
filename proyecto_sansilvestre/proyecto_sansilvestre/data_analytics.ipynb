# Convertir tiempos a segundos

```python
df["finish_seconds"] = df["finish_time"].dt.total_seconds()

df["year"] = df["race_date"]
```

```python
df[["finish_time", "finish_seconds"]].head()
```

# Performance Metrics 

Average finish times by age group 

```python
avg_time_by_age = df.groupby("age_group")["finish_seconds"].mean()
avg_time_by_age
```

Average finish times by gender 

```python
avg_time_by_gender = df.groupby("gender")["finish_seconds"].mean()
avg_time_by_gender
```

# Trend Analysis 

```python
# Extraer el año
df["year"] = pd.to_datetime(df["race_date"]).dt.year
```

Participation evolution over the years 

```python
participants_by_year = df.groupby("year").size()
participants_by_year
```

Male/female participation trends 

```python
gender_participation  = df.groupby(["year", "gender"]).size().unstack()
gender_participation
```

Evolution of mean and median times

```python
mean_median_by_year = df.groupby("year")["finish_seconds"].agg(["mean", "median"])
print(mean_median_by_year)
```

Minimum and maximum times over the years

```python
min_max_by_year = df.groupby("year")["finish_seconds"].agg(["min", "max"])
print(min_max_by_year)
```

# Top Performers 


Identify top performers by average time across all races they participated in



```python
top_runners = (
    df.groupby("runner_name")["finish_seconds"]
    .mean()
    .sort_values()
    .head(10)
)

top_runners
```

Summary report 

En esta fase del proyecto ("Data Analytics") se analizaron los datos de la carrera San Silvestre con el objetivo de extraer información relevante sobre el rendimiento de los participantes y la evolución del evento a lo largo del tiempo. 

El análisis de los tiempos medios por grupos de edad muestra diferencias claras entre categorías. Los grupos juveniles y senior masculinos presentan los mejores tiempos, mientras que las categorías de mayor edad o con limitaciones físicas registran tiempos más elevados, siguiendo patrones esperables en pruebas populares. En cuanto a la evolución temporal, la participación aumenta entre 2018 y 2019, desciende notablemente en 2021 (probablemente como consecuencia de la pandemia) y se recupera con fuerza a partir de 2023, alcanzando los valores más altos del periodo analizado. Aunque la participación masculina es mayor en todos los años, la participación femenina muestra una tendencia creciente, lo que indica una progresiva reducción de la brecha de género.

Esto sugiere que, aunque el nivel de los corredores más rápidos se mantiene constante, el evento ha incorporado un mayor número de participantes recreativos.

Finalmente, se identificaron los corredores con mejor rendimiento a partir de su tiempo medio en todas las carreras disputadas, destacando la regularidad como un factor clave del alto rendimiento. En conjunto, el análisis confirma que la San Silvestre es una prueba en crecimiento, cada vez más inclusiva, y con patrones de rendimiento coherentes con el perfil de sus participantes.
