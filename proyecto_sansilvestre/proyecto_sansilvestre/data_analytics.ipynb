# Convertir tiempos a segundos

```python
# 1. Aseguramos que sea string y extraemos solo la parte del tiempo (HH:MM:SS)
df['finish_time'] = df['finish_time'].astype(str).str.split(' ').str[-1]

# 2. Convertimos a Timedelta para que Pandas entienda que es una duración
df['finish_time'] = pd.to_timedelta(df['finish_time'])

# 3. Calculamos los segundos totales
df["finish_seconds"] = df["finish_time"].dt.total_seconds()

# 4. Aseguramos que el año sea numérico
df["year"] = pd.to_numeric(df["race_date"])

# Mostramos el resultado (Corregido: eliminada la coma extra y añadido runner_name para contexto)
print(df[["runner_name", "finish_time", "finish_seconds", "year"]].head())
```

# Performance Metrics 

Average finish times by age group 

```python
avg_time_by_age = df.groupby("age_group")["finish_seconds"].mean().sort_values(ascending=True)
avg_time_by_age
```

Average finish times by gender 

```python
avg_time_by_gender = df.groupby("gender")["finish_seconds"].mean()
avg_time_by_gender
```

# Trend Analysis 

```python
# Extraer el año
df["year"] = pd.to_numeric(df["race_date"], errors='coerce')
```

Participation evolution over the years 

```python
participants_by_year = df.groupby("year").size()
participants_by_year
```

Male/female participation trends 

```python
gender_participation  = df.groupby(["year", "gender"]).size().unstack()
gender_participation
```

Evolution of mean and median times

```python
mean_median_by_year = df.groupby("year")["finish_seconds"].agg(["mean", "median"])
print(mean_median_by_year)
```

Minimum and maximum times over the years

```python
min_max_by_year = df.groupby("year")["finish_seconds"].agg(["min", "max"])
print(min_max_by_year)
```

# Top Performers 


Identify top performers by average time across all races they participated in



```python
top_runners = (
    df.groupby("runner_name")["finish_seconds"]
    .mean()
    .sort_values()
    .head(10)
)

top_runners
```
